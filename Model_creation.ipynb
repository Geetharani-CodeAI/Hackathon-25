```python
import pandas as pd
dataset = pd.read_csv("crop_yield.csv")
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
independent = dataset[['Crop_Year', 'Area', 'Production', 'Annual_Rainfall', 'Fertilizer',
       'Pesticide','Crop_Bajra', 'Crop_Banana', 'Crop_Barley',
       'Crop_Black pepper', 'Crop_Cardamom', 'Crop_Cashewnut',
       'Crop_Castor seed', 'Crop_Coconut ', 'Crop_Coriander',
       'Crop_Cotton(lint)', 'Crop_Dry chillies', 'Crop_Garlic', 'Crop_Ginger',
       'Crop_Gram', 'Crop_Groundnut', 'Crop_Horse-gram', 'Crop_Jute',
       'Crop_Khesari', 'Crop_Maize', 'Crop_Moong(Green Gram)',
       'Crop_Niger seed', 'Crop_Oilseeds total', 'Crop_Onion',
       'Crop_Other  Rabi pulses', 'Crop_Other Cereals',
       'Crop_Other Kharif pulses', 'Crop_Other Summer Pulses',
       'Crop_Peas & beans (Pulses)', 'Crop_Potato', 'Crop_Ragi',
       'Crop_Rapeseed &Mustard', 'Crop_Rice', 'Crop_Safflower', 'Crop_Sesamum',
       'Crop_Small millets', 'Crop_Soyabean', 'Crop_Sugarcane',
       'Crop_Sunflower', 'Crop_Sweet potato', 'Crop_Tapioca', 'Crop_Tobacco',
       'Crop_Turmeric', 'Crop_Urad', 'Crop_Wheat', 'Crop_other oilseeds',
       'Season_Kharif     ', 'Season_Rabi       ', 'Season_Summer     ',
       'Season_Whole Year ', 'Season_Winter     ', 'State_Bihar',
       'State_Chhattisgarh', 'State_Delhi', 'State_Goa', 'State_Gujarat',
       'State_Himachal Pradesh', 'State_Jammu and Kashmir', 'State_Karnataka',
       'State_Kerala', 'State_Madhya Pradesh', 'State_Maharashtra',
       'State_Manipur', 'State_Puducherry', 'State_Punjab', 'State_State',
       'State_Tamil Nadu', 'State_Telangana', 'State_Uttar Pradesh',
       'State_Uttarakhand', 'State_West Bengal']]
```

```python
independent
```

```python
dependent = dataset[['Yield']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
weight = regressor.coef_
weight
```

```python
bias = regressor.intercept_
bias
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,Y_pred)
r_score
```

```python
# Support Vector Machine
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel='poly')
regressor.fit(X_train,Y_train)
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,Y_pred)
r_score
```

```python
# Decision Tree Regression
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_pred,Y_test)
r_score
```

```python
# Random Forest Regression
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100,random_state = 1)
regressor.fit(X_train,Y_train)
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_pred,Y_test)
r_score
```
